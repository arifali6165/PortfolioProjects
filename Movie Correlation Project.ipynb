```python
df=df.dropna()
```

```python
df.dtypes
```

```python
df['budget']=df['budget'].astype(int)
df['gross']=df['gross'].astype(int)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

**Visualization of Budget vs Gross**

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget For Flim')
plt.ylabel('Gross Earnings')
plt.show
```

```python
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

**Looking At Correlation**

```python
df.corr()
```

```python
correlation_metrix=df.corr()
sns.heatmap(correlation_metrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Correlation Features')
plt.ylabel('Movie Correlation Features')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.dtypes
```

```python
columns_to_encode=['name', 'rating', 'genre', 'released', 'director', 'writer', 'star', 'country', 'company']
for column in columns_to_encode:
  df[column] = le.fit_transform(df[column])
df
```

```python
df.dtypes
```

```python
correlation_metrix=df.corr()
sns.heatmap(correlation_metrix, annot=True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Correlation Features')
plt.ylabel('Movie Correlation Features')
```
