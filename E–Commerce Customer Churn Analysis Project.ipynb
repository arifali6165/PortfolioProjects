**ENCODING**

**GET DUMMY ENCODING**

```python
dummy=pd.get_dummies(df[['PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat']],drop_first=True)
dummy
```

```python
df1=pd.concat([df,dummy],axis=1)
df1
```

```python
df1=df1.drop(['CustomerID','MaritalStatus','PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat'],axis=1)
df1
```

```python
df1.dtypes
```

**MISSING VALUE HANDLING**

```python
df1.isna().sum()
```

```python
df1['Tenure']=df1['Tenure'].fillna(df1['Tenure'].mean())
```

```python
df1['WarehouseToHome']=df1['WarehouseToHome'].fillna(df1['WarehouseToHome'].mean())
```

```python
df1['HourSpendOnApp']=df1['HourSpendOnApp'].fillna(['HourSpendOnApp'].mean())
```

```python
df1['OrderAmountHikeFromlastYear']=df1['OrderAmountHikeFromlastYear'].fillna(df1['OrderAmountHikeFromlastYear'].mean())
```

```python
df1['CouponUsed']=df1['CouponUsed'].fillna(df1['CouponUsed'].mean())
```

```python
df1['OrderCount']=df1['OrderCount'].fillna(df1['OrderCount'].mean())
```

```python
df1['DaySinceLastOrder']=df1['DaySinceLastOrder'].fillna(df1['DaySinceLastOrder'].mean())
```

```python
df1.isna().sum()
```

**SPLITING INPUT AND OUTPUT VALUES**

```python
x=df1.drop(['Churn'],axis=1)
x
```

```python
y=df['Churn']
y
```

**TRAIN TEST POINT**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**NORMALIZATION**

**STANDARD SCALER**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

**MODLE CREATION**

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred1=LR.predict(x_test)
y_pred1
```

**PERFORMENCE EVALUATION**

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
```

```python
matrix=confusion_matrix(y_pred1,y_test)
matrix
```

```python
label=[1,0]
```

```python
mat=confusion_matrix(y_pred1,y_test)
cmd=ConfusionMatrixDisplay(mat,display_labels=label)
cmd.plot()
print(mat)
```

```python
score=accuracy_score(y_pred1,y_test)
score
```

```python
report=classification_report(y_pred1,y_test)
report
```

**MODLE CREATION**

**K NEAREST NEIGHBORS**

```python
KNN.fit(x_train,y_train)
```
